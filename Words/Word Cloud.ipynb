```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS
```

```python
stopwords = set(STOPWORDS)
```

```python
# open the file and read it into a variable alice_novel
alice_novel = open('./data/alice_novel.txt', 'r').read()
```

```python
alice_0 = WordCloud(
    background_color='white',
    max_words=2000,
    max_font_size=60,
    stopwords=stopwords ).generate(alice_novel);
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
plt.imshow(alice_0, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# adding do stopwords
stopwords.add('said')

alice_1 = WordCloud(
    background_color='white',
    max_words=2000,
    max_font_size=60,
    stopwords=stopwords ).generate(alice_novel);
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
plt.imshow(alice_1, interpolation='bilinear')
plt.axis('off')
#plt.savefig('alice_1.png')
plt.show()
```

```python
# read the mask image
# taken from
# http://www.stencilry.org/stencils/movies/alice%20in%20wonderland/255fk.jpg
alice_mask = np.array(Image.open("./data/alice_mask.png"))

alice_2 = WordCloud(background_color="white", 
                    max_words=2000, mask=alice_mask,
                    stopwords=stopwords, contour_width=3, 
                    contour_color='steelblue').generate(alice_novel);
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
plt.imshow(alice_2, interpolation='bilinear')
plt.axis("off")
plt.figure()

#alice_2.to_file("alice.png")
#plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
#plt.axis("off")
#plt.show()
```
